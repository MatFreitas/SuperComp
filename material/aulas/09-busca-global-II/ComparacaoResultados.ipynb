```python
import subprocess
import time
import os
import sys
import matplotlib.pyplot as plt
```

```python
def roda_com_entrada(executavel, arquivo_in):
    with open(arquivo_in) as f:
        start = time.perf_counter()
        proc = subprocess.run([executavel], input=f.read(), text=True, capture_output=True)
        end = time.perf_counter()

        # print('Saída:', proc.stdout)
        # print('Stderr:', proc.stderr)
        print('Tempo total(s):', end - start)

        return (proc.stdout, end - start)
```

```python
tempos_1 = []
tamanhos_1 = []
desempenhos_1 = []
for i in range(0, 4):
    with open(f'./in{i}.txt') as f:
        tamanhos_1.append(f.readline().strip().split(' ')[0])
    output, tempo = roda_com_entrada('./busca-global-recursiva', f'./in{i}.txt')
    desempenhos_1.append(output.strip()[-2:])
    tempos_1.append(tempo)

print(tamanhos_1, tempos_1, desempenhos_1)
plt.plot(tamanhos_1, tempos_1, color='green', linestyle='dashed')
plt.title('Busca Global Recursiva')
plt.xlabel('Tamanhos')
plt.ylabel('Tempos')
plt.show()
```
